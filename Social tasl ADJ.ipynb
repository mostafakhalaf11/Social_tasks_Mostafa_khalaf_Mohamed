```python
# direct graph
import networkx as nx
%matplotlib inline
```

```python
D = nx.DiGraph()
D.add_edges_from([(1,2),(2,3),(2,4),(3,2),(4,1),(6,3),(6,1)])
D.add_node(5)
nx.draw(D, with_labels=True, font_color='white', node_size=1000,node_color='brown')
```

```python
Adj = nx.adjacency_matrix(D).todense()
Adj
```

```python
from collections import Counter
import matplotlib.pyplot as plt

degree_sequence = [D.degree(n) for n in D.nodes]
degree_counts = Counter(degree_sequence)
min_degree, max_degree = min(degree_counts.keys()), max(degree_counts.keys())
plt.xlabel("Degree", fontsize=20)
plt.ylabel("Number of Nodes", fontsize=20)
plot_x = list(range(min_degree, max_degree + 1))
plot_y = [degree_counts.get(x, 0) for x in plot_x]
plt.bar(plot_x, plot_y)
```

```python
def plot_degree_dist(D):
    
    degrees = D.degree()
    degrees = dict(degrees)
    values = sorted(set(degrees.values()))
    print(values)
    histo = [list(degrees.values()).count(x) for x in values]
    P_k = [x / D.order() for x in histo]
    print(len(P_k))
    
    plt.figure()
    plt.bar(values, P_k)
    plt.xlabel("k",fontsize=40)
    plt.ylabel("p(k)", fontsize=40)
    plt.title("Degree Distribution", fontsize=40)
    
    plt.show()
    plt.figure()
    plt.grid(False)
    plt.loglog(values, P_k, "bo")
    plt.xlabel("k", fontsize=40)
    plt.ylabel("log p(k)", fontsize=40)
    plt.title("log Degree Distribution")
    plt.show()
    plt.show()
```

```python
plot_degree_dist(D)
```
